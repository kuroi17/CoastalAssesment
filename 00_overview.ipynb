## 🎯 Project Goal

Develop an **automated system** to monitor coastal environmental changes and provide **evidence-based recommendations** to government agencies for policy-making and resource allocation.

### Key Objectives:
1. ✅ Detect **seagrass loss** (marine ecosystem degradation)
2. ✅ Monitor **shoreline erosion** (sand vs water classification)
3. ✅ Quantify **area changes** over time (km²)
4. ✅ Generate **actionable reports** for decision-makers

---

## 🛰️ Data Source

**Copernicus Sentinel-2 Level 2A**
- 🆓 **Free and open** satellite data
- 📅 Revisit every **5 days**
- 📏 **10m resolution** (very detailed)
- 🎨 **13 spectral bands** (we use 4)

### Bands Used:

| Band | Name | Wavelength | Use Case |
|------|------|------------|----------|
| **B02** | Blue | 490 nm | Water body detection |
| **B03** | Green | 560 nm | Vegetation health |
| **B04** | Red | 665 nm | Vegetation stress |
| **B08** | NIR | 842 nm | Vegetation density |

---

## 🧠 Technical Approach

### Phase 1: Feature Engineering
**NDVI (Normalized Difference Vegetation Index)**
$$
\text{NDVI} = \frac{\text{NIR} - \text{Red}}{\text{NIR} + \text{Red}}
$$

**Why NDVI?**
- Healthy vegetation reflects **more NIR**, absorbs **more Red** → High NDVI
- Water/sand reflects **less NIR** → Low NDVI
- Perfect for detecting seagrass vs water vs sand

### Phase 2: Deep Learning Classification
**1D Convolutional Neural Network (CNN)**
- Input: 5 features (B02, B03, B04, B08, NDVI)
- Output: 4 classes (Cloud, Seagrass, Water, Sand)
- Why 1D CNN? Analyzes **spectral signatures** (pixel-level classification)

### Phase 3: Analysis & Reporting
- Calculate area in km² (pixel size = 10m × 10m)
- Multi-year trend analysis
- Generate visualizations for stakeholders
